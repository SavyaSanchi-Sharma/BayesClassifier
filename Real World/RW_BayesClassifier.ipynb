```python
import pandas as pd
import BayesClassifier as bc
import accuracy as ac
```

```python
class1_train=pd.read_csv('class1Train.csv')
class2_train=pd.read_csv('class2Train.csv')
class3_train=pd.read_csv('class3Train.csv')
class1_test=pd.read_csv('class1Test.csv')
class2_test=pd.read_csv('class2Test.csv')
class3_test=pd.read_csv('class3Test.csv')
```

```python
class1_train['Class']=1
class2_train['Class']=2
class3_train['Class']=3
class1_test['Class']=1
class2_test['Class']=2
class3_test['Class']=3
```

```python
df_train=pd.concat([class1_train,class2_train,class3_train],axis=0).drop('Unnamed: 0',axis=1)
df_train
```

```python
df_test=pd.concat([class1_test,class2_test,class3_test],axis=0).drop('Unnamed: 0',axis=1)
df_test
```

```python
y_true=df_test['Class']
```

```python
classifier =bc.bc(df=df_train, class_column='Class', features=['x1', 'x2'])
y_hat=classifier.predict(df_test[['x1', 'x2']])
y_hat
```

```python
acc=ac.accuracy(y_true,y_hat)
acc.confusion_matrix()
```

```python
acc.accu_percent()
```

```python
acc.precision()
```

```python
acc.recall()
```

```python
acc.f_score(acc.precision())
```
